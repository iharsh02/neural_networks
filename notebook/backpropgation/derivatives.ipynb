```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
# scaler function 
# quadratic equation , differentiate it to get the slope at any point
def f(x):
    return 3*x**2 - 4*x + 5
```

```python

f(3.0)
```

```python
# giving range of values to plot the parabolic curve
xs = np.arange(-5.0, 5.0, 0.1)
ys = f(xs)
print(f"xs: {xs[:5]}... ys: {ys[:5]}")
plt.plot(xs, ys)
```

```python
h =  0.001
x = 3.0

(f(x + h) - f(x))/h


```

```python
a = 2.0
b = -3.0
c = 10.0

d = a*b + c
print(d)
```

```python
h=0.001

a = 2.0
b = -3.0
c = 10.0

d1 = a*b + c

a += h

d2 = a*b + c

print(f" d1 = {d1} , d2 = {d2}")
print("slope: ", (d2-d1)/h )
```
